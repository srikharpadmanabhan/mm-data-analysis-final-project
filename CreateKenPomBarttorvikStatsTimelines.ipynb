```python
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
```

```python
# Read ken pom into dataframe
stats_df = pd.read_csv("MM Data/KenPom Barttorvik.csv")
stats_df
```

```python

# First png will be aplot of BADJ O and BADJ D over time

badjO_yearly = stats_df.groupby("YEAR")["BADJ O"].mean().reset_index()

badjD_yearly = stats_df.groupby("YEAR")["BADJ D"].mean().reset_index()


badjO_yearly = badjO_yearly.sort_values('YEAR')
badjD_yearly = badjD_yearly.sort_values('YEAR')

first_year = badjO_yearly['YEAR'].min() # Could also use badjD yearly


badjO_yearly["YEAR"] = badjO_yearly["YEAR"] - first_year
badjD_yearly["YEAR"] = badjD_yearly["YEAR"] - first_year

plt.plot(badjO_yearly['YEAR'], badjO_yearly['BADJ O'], label='BADJ O')
plt.plot(badjD_yearly['YEAR'], badjD_yearly['BADJ D'], label='BADJ D')

# Customize plot
plt.xlabel(f'Year starting from {first_year}')
plt.ylabel('Yearly mean of badj (O/D)')
plt.title('Aggregated Mean of BADJ')
plt.grid(True)
plt.legend()

plt.savefig("annual-stats-averages/BADJ-annual.png")

plt.clf()


```

```python
# Now we can look at KADJ O and KADJ D

kadjO_yearly = stats_df.groupby("YEAR")["KADJ O"].mean().reset_index()

kadjD_yearly = stats_df.groupby("YEAR")["KADJ D"].mean().reset_index()


kadjO_yearly = kadjO_yearly.sort_values('YEAR')
kadjD_yearly = kadjD_yearly.sort_values('YEAR')

first_year = kadjO_yearly['YEAR'].min() # Could also use kadjD yearly


kadjO_yearly["YEAR"] = kadjO_yearly["YEAR"] - first_year
kadjD_yearly["YEAR"] = kadjD_yearly["YEAR"] - first_year

plt.plot(kadjO_yearly['YEAR'], kadjO_yearly['KADJ O'], label='KADJ O')
plt.plot(kadjD_yearly['YEAR'], kadjD_yearly['KADJ D'], label='KADJ D')

# Customize plot
plt.xlabel(f'Year starting from {first_year}')
plt.ylabel('Yearly mean of KADJ (O/D)')
plt.title('Aggregated Mean of KADJ')
plt.grid(True)
plt.legend()

plt.savefig("annual-stats-averages/KADJ-annual.png")

plt.clf()
```

```python
# Now we can look at K OFF and K DEF

kOff_yearly = stats_df.groupby("YEAR")["K OFF"].mean().reset_index()

kDef_yearly = stats_df.groupby("YEAR")["K DEF"].mean().reset_index()


kOff_yearly = kOff_yearly.sort_values('YEAR')
kDef_yearly = kDef_yearly.sort_values('YEAR')

first_year = kOff_yearly['YEAR'].min() 


kOff_yearly["YEAR"] = kOff_yearly["YEAR"] - first_year
kDef_yearly["YEAR"] = kDef_yearly["YEAR"] - first_year

plt.plot(kOff_yearly['YEAR'], kOff_yearly['K OFF'], label='K OFF')
plt.plot(kDef_yearly['YEAR'], kDef_yearly['K DEF'], label='K DEF')

# Customize plot
plt.xlabel(f'Year starting from {first_year}')
plt.ylabel('Yearly mean of K (OFF/DEF)')
plt.title('Aggregated Mean of K')
plt.grid(True)
plt.legend()

plt.savefig("annual-stats-averages/K-annual.png")

plt.clf()
```

```python
# Now let's look at EFG, 2PT and 3PT
shooting_yearly = stats_df.groupby("YEAR")[["EFG%", "2PT%", "3PT%"]].mean().reset_index()


shooting_yearly = shooting_yearly.sort_values('YEAR')
first_year = shooting_yearly['YEAR'].min() 
shooting_yearly["YEAR"] = shooting_yearly["YEAR"] - first_year

for stat in ["EFG%","2PT%","3PT%"]:
    plt.plot(shooting_yearly['YEAR'], shooting_yearly[stat], label=stat)


# Customize plot
plt.xlabel(f'Year starting from {first_year}')
plt.ylabel("Yearly mean of Shooting %'s")
plt.title('Aggregated Average Shooting%')
plt.grid(True)
plt.legend()

plt.savefig("annual-stats-averages/shooting-percentages-annual.png")

plt.clf()
```

```python
# Lets look at TOV%, OREB%, DREB%, BLK%, and AST
tov_and_reb_yearly = stats_df.groupby("YEAR")[["TOV%", "OREB%", "DREB%", "BLK%", "AST%"]].mean().reset_index()

tov_and_reb_yearly = tov_and_reb_yearly.sort_values('YEAR')
first_year = tov_and_reb_yearly['YEAR'].min() 
tov_and_reb_yearly["YEAR"] = tov_and_reb_yearly["YEAR"] - first_year

for stat in ["TOV%", "OREB%", "DREB%", "BLK%", "AST%"]:
    plt.plot(tov_and_reb_yearly['YEAR'], tov_and_reb_yearly[stat], label=stat)


# Customize plot
plt.xlabel(f'Year starting from {first_year}')
plt.ylabel("Yearly mean of TOV/REB %'s")
plt.title('Aggregated Average Rebounding and Turnovers')
plt.grid(True)
plt.legend()

plt.savefig("annual-stats-averages/rebounding-assist-tov-and-blk-percentages-annual.png")

plt.clf()
```
