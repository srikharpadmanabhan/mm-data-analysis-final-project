```python
import pandas as pd
import pickle
```

```python
# Loading 2024 games
with open('final_data_pickles/games-without-scores.pkl', 'rb') as games_file:
    games_2024 = pickle.load(games_file)

len(games_2024)
```

```python
games_stats_2024 = [game.get_game_stats_dict() for game in games_2024]
games_stats_2024[0]
```

```python


games_2024_df = pd.DataFrame.from_dict(games_stats_2024, orient="columns")
```

```python
games_2024_df.head()
```

```python
# Get top 15 list from pickle
with open('TOP_STATS_LIST.pkl', 'rb') as file:
    top15_list = pickle.load(file)
```

```python
games_subset_2024 = games_2024_df[top15_list]
```

```python
games_subset_2024
```

```python
# Load model 4 and model 3
from tensorflow.keras.models import load_model

logistic_model = load_model("saved-models/model-3.h5")
linear_model = load_model("saved-models/model-4-linear.h5")

# Load scaler
with open('MinMaxScaler.pkl', 'rb') as file:
    scaler = pickle.load(file)


```

```python
X = scaler.transform(games_subset_2024)
X.shape
```

```python
linear_predictions = linear_model.predict(X)
logistic_predictions = logistic_model.predict(X)
```

```python
len(linear_predictions), len(logistic_predictions)
```

```python
team1_names = games_2024_df["TEAM 1 Name"]
team2_names = games_2024_df["TEAM 2 Name"]

combined = list(zip(team1_names, team2_names, linear_predictions, logistic_predictions))
```

```python
# Convert prediction arrays to float
combined = [(a, b, c[0], d[0]) for a, b, c, d in combined]
```

```python
linear_correct, logistic_correct, total = 0, 0, 0
for game in combined:
    result = correct_results[(game[0], game[1])]
    if result == "Skip":
        continue
    
    total += 1
    if result == "Win":
        if game[2] > 0:
            linear_correct += 1
        if game[3] > 0.5:
            logistic_correct += 1
    elif result == "Loss":
        if game[2] < 0:
            linear_correct += 1
        if game[3] < 0.5:
            logistic_correct += 1


print(f"Linear Accuracy: {linear_correct/total}, Logistic Accuracy: {logistic_correct / total}")
    
        
```
