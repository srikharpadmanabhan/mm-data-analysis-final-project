```python
import pandas as pd
import pickle

pd.set_option('display.max_columns', 244)
```

```python
# Loading game list
with open('final_data_pickles/games-with-scores.pkl', 'rb') as games_file:
    game_data = pickle.load(games_file)

len(game_data)
```

```python
# Convert list of games to list of dictionaries (game_stats_dicts)

game_stats_list = [game.get_game_stats_dict() for game in game_data]

print(game_stats_list[:3])
print(len(list(game_stats_list[0].keys())))
```

```python
games_df = pd.DataFrame.from_dict(game_stats_list, orient='columns')
```

```python
games_df
```

```python
games_df.head()
```

```python
# Get rows with nan values

nan_rows = games_df.isna().any(axis=1).sum()

print(f'Number of rows with NaN or None values: {nan_rows}')
```

```python
# Find which columns have Nans

cols_with_nan = list(games_df.columns[games_df.isna().any()])

print("Columns with nans:", len(cols_with_nan))
print(cols_with_nan)

print([col for col in list(games_df.columns) if col not in cols_with_nan])
```

```python
# Dropping rows that have nan values. This will basically be a set of ~440 rows that have all the data (ie shooting splits, 538 rating)
games_with_all_stats = games_df.dropna(axis=0)

# Dropping columns with NaN values, this will basically be 944 rows with a subset of the columns such that all rows don't have nan
games_nan_cols_dropped = games_df.dropna(axis=1)
```

```python
games_with_all_stats.tail()
```

```python
games_nan_cols_dropped.tail()
```

```python
# Save files to csv, no need for an index and float_format is to round to three places after decimal
games_with_all_stats.to_csv("csv_files/games-with-all-stats.csv", index=False, float_format='%.3f')
games_nan_cols_dropped.to_csv("csv_files/all-games-with-annual-stats.csv", index=False, float_format='%.3f')
```
