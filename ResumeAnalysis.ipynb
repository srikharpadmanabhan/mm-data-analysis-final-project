```python
import pandas as pd
import pickle
```

```python
# Load Resume.csv

resume_df = pd.read_csv("MM Data/Resumes.csv")
resume_df.head()
```

```python
# create overall dictionary for team resumes
resume_dict = dict()

# Ignore these fields (Q1 + Q2 is unnecessary because we already ahve access to Q1 and Q2 W)
IGNORE_FIELDS_RESUME = ["YEAR","TEAM NO","TEAM","SEED","ROUND","Q1 PLUS Q2 W","BID TYPE"]



# Iterate through df, and create the conferences from the Conferences.py file ignoring the fields above
for _, row in resume_df.iterrows():
    row = row.to_dict()
    
    # Get the right fields for the stats, ignoring the id variables above
    resume_team_stats = {
        ("(Resumes) " + k): v for k, v in row.items() if k not in IGNORE_FIELDS_RESUME
    }
    
    resume_dict[(row["TEAM"], row["YEAR"])] = resume_team_stats

    
# Dump data into pickle file
with open('pickle_files/resume-data.pkl', 'wb') as resume_file:
    pickle.dump(resume_dict, resume_file, protocol=pickle.HIGHEST_PROTOCOL)
    
print(list(resume_dict.items())[0])
```
