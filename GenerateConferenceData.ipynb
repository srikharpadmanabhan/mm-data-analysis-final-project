```python
import pandas as pd
import pickle
```

```python
conference_stats_df = pd.read_csv("MM Data/Conference Stats.csv")
```

```python
conference_stats_df.head()
```

```python
from Conference import Conference


IGNORE_FIELDS_FOR_CONFERENCE_STATS = ["YEAR","CONF ID","CONF","W", "L"]

# create overall conference dictionary 
conference_data_dict = dict()



# Iterate through df, and create the conferences from the Conferences.py file ignoring the fields above
for _, row in conference_stats_df.iterrows():
    row = row.to_dict()
    
    # Get the right fields for the stats, ignoring the id variables above
    curr_conference_stats = {
        ("(Conference) " + k): v for k, v in row.items() if k not in IGNORE_FIELDS_FOR_CONFERENCE_STATS
    }
    
    conference_data_dict[(row["CONF"], row["YEAR"])] = Conference(conference_name=row["CONF"], conference_stats=curr_conference_stats)

    
# Dump data into pickle file
with open('pickle_files/conference-data.pkl', 'wb') as conference_file:
    pickle.dump(conference_data_dict, conference_file, protocol=pickle.HIGHEST_PROTOCOL)
    
print(list(conference_data_dict.items())[0])
print((list(conference_data_dict.items())[0])[1])
```
