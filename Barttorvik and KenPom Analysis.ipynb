```python
# Open the three Battorvi

barttorvik_df = pd.read_csv("MM Data/Barttorvik Away-Neutral.csv")
kenpom_with_barttorvik = pd.read_csv("MM Data/KenPom Barttorvik.csv")
```

```python
# create overall dictionary to hold team data from kenpom and barttorvik  data
combined_kenpom_dict = dict()



# Store csv in dictionary and save as pkl file
for _, row in kenpom_with_barttorvik.iterrows():
    row = row.to_dict()
    
    combined_kenpom_dict[(row["TEAM"], row["YEAR"])] = {
        k: v for k, v in row.items() if "RANK" in k
    }

    
    

# Dump data into pickle file
with open('pickle_files/kenpom-barttorvik-combined.pkl', 'wb') as single_season_team_file:
    pickle.dump(combined_kenpom_dict, single_season_team_file, protocol=pickle.HIGHEST_PROTOCOL)
    
```
