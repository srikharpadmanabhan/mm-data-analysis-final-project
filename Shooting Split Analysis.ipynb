```python
import pandas as pd
import pickle
```

```python
shooting_splits_df = pd.read_csv("MM Data/Shooting Splits.csv")

shooting_splits_df.head(7)
```

```python
FIELDS_TO_IGNORE = ["YEAR","TEAM NO","TEAM ID","TEAM","CONF"]

# create shooting splits dictionary
shooting_splits_dict = dict()



# Store data for shooting splits in dictionary and save as pkl file
for _, row in shooting_splits_df.iterrows():
    row = row.to_dict()
    
    # Get the shooting stats by onl;y looling at non-ranking stats and fields not in ignored list above
    curr_shooting_stats = {
        ("(Shooting Splits) " + k): v for k, v in row.items() if k not in FIELDS_TO_IGNORE and "RANK" not in k
    }
    
    
    shooting_splits_dict[(row["TEAM"], row["YEAR"])] = curr_shooting_stats
    

    
# Dump data into pickle file
with open('pickle_files/shooting-splits-data.pkl', 'wb') as shooting_splits_file:
    pickle.dump(shooting_splits_dict, shooting_splits_file, protocol=pickle.HIGHEST_PROTOCOL)
    
print((list(shooting_splits_dict.items()))[:3])
print(len(shooting_splits_dict), (list(shooting_splits_dict.items())[0][1]))
```
